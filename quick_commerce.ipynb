```python
#handling missing values
```

```python
df.isnull().sum()
```

```python
df['City']=df['City'].fillna('Unknown')
```

```python
df['Items_Count'] = df.groupby('Product_Category')['Items_Count'].transform(lambda x: x.fillna(x.median()))
```

```python
df['rating_missing'] = df['Customer_Rating'].isnull().astype(int)
```

```python
df['partner_rating_missing'] = df['Delivery_Partner_Rating'].isnull().astype(int)
```

```python
#feature engineering
```

```python
df['age_group'] = pd.qcut(
    df['Customer_Age'], 
    q=3, 
    labels=['young_adult', 'middle-aged', 'senior'],
    duplicates='drop'
)
```

```python
df['is_late'] = (df['Delivery_Time_Min'] > 30).astype(int)
```

```python
df['Customer_Rating'] = df['Customer_Rating'].round(1)
df['Delivery_Partner_Rating'] = df['Delivery_Partner_Rating'].round(1)
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Prafulsql4266@localhost/quick_commerce")

df.to_sql('quick_commerce', con=engine, if_exists='replace', index=False)
```
